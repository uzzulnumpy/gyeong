```python
a.drop(['2012년 이전', '2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년'], axis=1, inplace=True)
```

```python
a.tail()
```

```python
a['총계'] = a['총계'].apply(lambda x: x.replace(',', ''))
a['2019년'] = a['2019년'].apply(lambda x: x.replace(',', ''))
a['2020년'] = a['2020년'].apply(lambda x: x.replace(',', ''))
a['2021년'] = a['2021년'].apply(lambda x: x.replace(',', ''))
a = a.astype({'총계': 'int', '2019년': 'int', '2020년': 'int',
             '2021년': 'int'})
a.info()
```

```python
a['증가율'] = (a['2019년'] + a['2020년'] + a['2021년'] / a['총계'] * 100)
a.sort_values(by='증가율', ascending=False).head()
```

```python
a.drop(0, axis=0, inplace=True)
```

```python
a
```

```python
b.drop(0, axis=0, inplace=True)
```

```python
b['구분'].unique()
```

```python
b.isnull().sum()
```

```python
b['외국인비율'] = b['외국인'] / b['인구수'] * 100
b['고령자비율'] = b['고령자'] / b['인구수'] * 100
b.head()
```

```python
data_result = pd.merge(a, b, on='구분')
data_result.set_index('구분', inplace=True)
data_result.head()
```

```python
data_result.corr()
```

```python
data_result['총계'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
data_result['cctv비율'] = data_result['총계'] / data_result['인구수'] * 100
data_result['cctv비율'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.show()
```

```python
plt.figure()
plt.scatter(data_result['인구수'], data_result['총계'], s=50)
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.title('산점도 찍어볼게')
plt.grid()
plt.show()
```
